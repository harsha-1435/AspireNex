# Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('IRIS.csv')
df.info()
```

```python
df.isnull().sum()
```

```python
# Finding duplicates and removing
df[df.duplicated(keep=False)]
```

```python
df[df.duplicated()]
```

```python
duplicates = df.duplicated()
df = df[~duplicates]
```

```python
df[df.duplicated()]
```

```python
df.info()
```

```python
df.drop('species', axis=1).corr()
```

```python
# Map species names to numeric labels
species_mapping = {"Iris-setosa": 1, "Iris-versicolor": 2, "Iris-virginica": 3}
df["species"] = df["species"].replace(species_mapping)

# correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# IQR method
def find_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers

outliers = df.iloc[:, :-1].apply(find_outliers_iqr)
outliers_count = outliers.sum()

outliers_count
```

```python
# Boxplot to visualize outliers in sepal_width
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(y=df["sepal_width"])
plt.title("Box Plot of Sepal Width")
plt.ylabel("Sepal Width")
plt.show()
```

```python
# Average values before removing outliers
average_before = df.iloc[:, :-1].mean()

df_without_outliers = df[~outliers.any(axis=1)]

# Average values after removing outliers
average_after = df_without_outliers.iloc[:, :-1].mean()
average_comparison = pd.DataFrame({
    'Before Removing Outliers': average_before,
    'After Removing Outliers': average_after
})

average_comparison
```

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

averages = df.groupby('species')[['sepal_length', 'sepal_width']].mean().reset_index()
plt.figure(figsize=(8, 4))
sns.barplot(data=averages.melt(id_vars='species'), x='species', y='value', hue='variable')
plt.title('Average Sepal Length and Average sepal Width by Species')
plt.ylabel('Average (cm)')
plt.xlabel('Species')
plt.legend(title='Measurement')
plt.show()
```

```python
# Average petal length and petal width by species
average=df.groupby('species')[['petal_length','petal_width']].mean().reset_index()
plt.figure(figsize=(8,4))
sns.barplot(data=average.melt(id_vars='species'), x='species', y='value', hue='variable')
plt.title('Average petal_length and Average petal_width by species')
plt.xlabel('species')
plt.ylabel('Average (cm)')
plt.legend(title='Measurement')
plt.show()
```

```python
#Distribution of Iris Species
count = df['species'].value_counts()
plt.figure(figsize=(8, 4))
count.plot.pie(autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribution of Iris Species')
plt.ylabel('')
plt.show()
```

```python
# Scatter plot to visualize the distribution of Measures 
copy_df=pd.read_csv('IRIS.csv')

fig, axes = plt.subplots(2, 2, figsize=(15, 12))
sns.scatterplot(data=copy_df, x='sepal_length', y='sepal_width', hue='species', palette='Set1', ax=axes[0, 0])
axes[0, 0].set_title('Scatter Plot of Sepal Length vs Sepal Width')
axes[0, 0].set_xlabel('Sepal Length (cm)')
axes[0, 0].set_ylabel('Sepal Width (cm)')

sns.scatterplot(data=copy_df, x='petal_length', y='petal_width', hue='species', palette='Set1', ax=axes[0, 1])
axes[0, 1].set_title('Scatter Plot of Petal Length vs Petal Width')
axes[0, 1].set_xlabel('Petal Length (cm)')
axes[0, 1].set_ylabel('Petal Width (cm)')

sns.scatterplot(data=copy_df, x='petal_length', y='sepal_length', hue='species', palette='Set1', ax=axes[1, 0])
axes[1, 0].set_title('Scatter Plot of Petal Length vs Sepal Length')
axes[1, 0].set_xlabel('Petal Length (cm)')
axes[1, 0].set_ylabel('Sepal Length (cm)')

sns.scatterplot(data=copy_df, x='sepal_width', y='petal_width', hue='species', palette='Set1', ax=axes[1, 1])
axes[1, 1].set_title('Scatter Plot of Sepal Width vs Petal Width')
axes[1, 1].set_xlabel('Sepal Width (cm)')
axes[1, 1].set_ylabel('Petal Width (cm)')

plt.tight_layout()
plt.show()
```
